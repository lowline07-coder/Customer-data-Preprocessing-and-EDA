# 4. Tensors

## What are Tensors?

A Tensor is a mathematical structure used to store numerical data in machine learning and deep learning.

Tensors are generalized forms of arrays.

## Types of Tensors

| Tensor Type         | Example                 |
| ------------------- | ----------------------- |
| Scalar (0D)         | Single value            |
| Vector (1D)         | List of values          |
| Matrix (2D)         | Rows and columns        |
| Higher Order Tensor | Multi-dimensional array |

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import requests
```

## 2. Load CSV File

```python
csv_data = pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\projects\customer_data.csv")

print(csv_data.head())
```

## 3. Parse JSON File

```python
csv_data.to_json(
    r"C:\Users\Lenovo\OneDrive\Desktop\projects\customer_data.csv",
    orient='records'
)
```

```python
json_data = pd.read_json(
    r"C:\Users\Lenovo\OneDrive\Desktop\projects\customer_data.csv"
)

print(json_data.head())
```

## 4. Connect to SQL Database

```python
import sqlite3

connection = sqlite3.connect(
    'customer_database.db'
)

csv_data.to_sql(
    'customers',
    connection,
    if_exists='replace',
    index=False
)
```

```python
sql_data = pd.read_sql(
    'SELECT * FROM customers',
    connection
)

print(sql_data.head())
```

## 5. Fetch Data from API

```python
url = 'https://jsonplaceholder.typicode.com/users'

response = requests.get(url)

api_data = pd.DataFrame(
    response.json()
)

print(api_data.head())
```

## 6. Merge Data Sources

```python
merged_data = pd.concat(
    [csv_data, json_data],
    ignore_index=True
)

print(merged_data.head())
```

## PART C – DATA UNDERSTANDING & CLEANING

```python
print(merged_data.head())
```

```python
print(merged_data.info())
```

```python
print(merged_data.describe())
```

```python
print(
    merged_data.isnull().sum()
)
```

```python
merged_data['Income'].fillna(
    merged_data['Income'].mean(),
    inplace=True
)
```

```python
merged_data.drop_duplicates(
    inplace=True
)
```

```python
merged_data['Age'] = merged_data[
    'Age'
].astype(int)
```

```python
merged_data.drop(
    columns=['Customer_ID'],
    inplace=True
)
```

## PART D – EXPLORATORY DATA ANALYSIS (EDA)
